## 03 Merging Data

```python
#dropping '_merge' column from df_ords_prods_combined
df_ords_prods_combined = df_ords_prods_combined.drop(columns = ['_merge'])
```

```python
#checking df_ords_prods_combined
df_ords_prods_combined.head()
```

```python
#merging df_prods and df_ords_prods_combined on 'order_id' column
ords_prods_merged = df_prods.merge(df_ords_prods_combined, on = 'product_id', indicator = True)
```
